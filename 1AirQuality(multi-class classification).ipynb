```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelEncoder = LabelEncoder()
```

```python
df
```

```python
df["Air Quality Category"] = labelEncoder.fit_transform(df["Air Quality Category"])
```

```python
df["Geo Place Name"] = labelEncoder.fit_transform(df["Geo Place Name"])
```

```python
df["Time Period"] = labelEncoder.fit_transform(df["Time Period"])
```

```python
df = df.drop("Start_Date" , axis =1 )
```

```python
df["Air Quality Category"].unique()
```

```python
y = df["Air Quality Category"].values
x = df.drop("Air Quality Category" , axis=1 ).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y , test_size=0.12 , random_state=16 )
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Activation , Dropout
from sklearn.metrics import classification_report  , confusion_matrix
```

```python
modelTensor = Sequential()

modelTensor.add(Dense(3 , activation="relu" , input_shape=(x_train.shape[1],)))
modelTensor.add(Dense(16 , activation="relu"))
modelTensor.add(Dense(4 , activation="relu"))

modelTensor.add(Dense(3 , activation="softmax"))

modelTensor.compile(optimizer="adam" , loss="sparse_categorical_crossentropy" , metrics=["accuracy"])
```

```python
modelTensor.fit(x= x_train , y= y_train , epochs=70 , validation_data=(x_test, y_test) , verbose=1 )
```

```python
lossData = pd.DataFrame(modelTensor.history.history)
```

```python
lossData.plot()
```

```python
Y_pred= modelTensor.predict(x_test)
```

```python
Y_pred.shape
```

```python
predicted_classes = np.argmax(Y_pred , axis=1)
```

```python
confussionMat = (confusion_matrix(y_test , predicted_classes))
```

```python
sb.heatmap(confussionMat, annot=True, fmt='d', cmap='Blues', xticklabels=[0, 1, 2], yticklabels=[0, 1, 2])
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
print(classification_report(y_test , predicted_classes))
```
